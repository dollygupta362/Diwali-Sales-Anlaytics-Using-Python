```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("Diwali Sales Data.csv")
df.shape
```

```python
df = pd.read_csv("Diwali Sales Data.csv", encoding='unicode_escape')
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status', 'Unnamed1'], axis=1, inplace=True, errors='ignore')
```

```python
df.info()
```

```python
df.drop(['unnamed1'], axis=1, inplace=True)
```

```python
pd.isnull(df)
```

```python
#check for null values
pd.isnull(df).sum()
```

```python
#drop null values
df.dropna(inplace=True)
```

```python
#Change data type
df['Amount'] = df['Amount'].astype('int')
df['Amount'].dtypes
```

```python
df.columns
```

```python
#describe()method return discription of the data in the DataFrame (i.e. count, mean, std, etc)
df.describe()
```

```python
# use describe()for specific columns
df[['Age','Orders', 'Amount']].describe()
```

Exploratory Data Analysis


Gender

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
                                                                         
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x ='Gender',y= 'Amount',data=sales_gen)
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

Age

```python
df.columns
```

```python
ax = sns.countplot (data = df, x = 'Age Group', hue= 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot (x = 'Age Group',y= 'Amount', data = sales_age)
```

```python
From above graphs we can seee that most of the buyer are of age group between 26-35 yrs female
```

State


```python
df.columns
```

```python
#Total number of orders from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
                                                                                                                   
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                           
sns.set(rc={'figure.figsize': (15, 5)})
sns.barplot(data=sales_state, x='State', y='Orders')
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize': (15, 5)})
sns.barplot(data=sales_state, x='State', y='Amount')
```

```python
df.columns
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize': (6, 5)})
sns.barplot(data=sales_state, x='Marital_Status', y='Amount', hue='Gender')
plt.title("Total Purchase Amount by Marital Status and Gender")
plt.show()
```

```python
From above graphs wse can see that most of the buyer are married women and they have high purchisang power
```

```python
Occupation
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize': (20, 5)})
ax = sns.countplot(data=df, x='Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation=45)
plt.title('Customer Count by Occupation')
plt.show()

```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
                                                                                                                   
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                           
sns.set(rc={'figure.figsize': (20, 5)})
sns.barplot(data=sales_state, x='Occupation', y='Amount')
```

```python
From above graph we can see that most of the buyer are working in IT sector and healthcare aviation sector
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize': (20, 5)})

ax = sns.countplot(data=df, x='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)

plt.xticks(rotation=45)
plt.title('Count of Orders by Product Category')
plt.show()
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
                                                                                                                   
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                           
sns.set(rc={'figure.figsize': (20, 5)})
sns.barplot(data=sales_state, x='Product_Category', y='Amount')
```

```python
From above graph we can see thatmost of the sold product are from food footwear and electronics category
```

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
                                                                                                                   
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                           
sns.set(rc={'figure.figsize': (20, 5)})
sns.barplot(data=sales_state, x='Product_ID', y='Orders')
```

Conclusion:
The Diwali Sales Data analysis indicates that marketing efforts should target women, especially married individuals aged 26-35, focusing on key states like Uttar Pradesh, Maharashtra, and Karnataka.
